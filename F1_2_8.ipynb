Primero, tomemos el sistema de referencia como muestra la siguiente figura:

¿Porque tomamos este sistema de referencia?.

Basicamente, sabemos que no hay rozamiento entre el plano y el proyectil y tampoco entre el plano y la varilla. Además, de esta forma la varilla solo se mueve en el plano $(\hat{y}, \hat{z})$ y el movimiento del proyectil solo ocurre en el plano $(\hat{x}, \hat{y})$. Luego, la dinamica de ambos cuerpos esta regida por las mismas fuerzas en las mismas componentes, donde tenemos:

$$Proyectil \left\{ \begin{array}{ccc}
  \sum F_x : & - &\\
  \sum F_y : & P_y = m_p \ddot{y}_p & (1)\\
  \sum F_z : & N + P_z = m_p \ddot{z}_p & (2)
 \end{array}\right.$$

 Donde $\overline{P}_p=(0, P_y, P_z) = - m_p g \cdot (0, \sin{\alpha}, \cos{\alpha})$

 Luego, como el proyectil no se despega del plano, entonces  $\ddot{z}_p = 0$.

 Ahora, para la varilla tenemos:

$$Varilla \left\{ \begin{array}{ccc}
  \sum F_x : & - &\\
  \sum F_y : & P_y = m_v \ddot{y}_v & (3)\\
  \sum F_z : & N + P_z = m_v \ddot{z}_v & (4)
 \end{array}\right.$$

 Donde $\overline{P}_v=(0, P_y, P_z) = - m_v g \cdot (0, \sin{\alpha}, \cos{\alpha})$

 Luego, como la varilla no se despega del plano, entonces  $\ddot{z}_v = 0$.

 Parece redundante, pero, las ecuaciónes dinamicas son identicas, salvo por las masas, pero sus aceleraciones tienen formas análogas. 
 
 Ahora, esto nos garantiza que su cinematica sea identica, es más, la varilla parte del reposo y siempre se mueve sobre $\hat{y}$. En cámbio, el proyectil se mueve sobre el plano $(\hat{x}, \hat{y})$ y en particular parde con un $\overline{v}_0 \neq 0$, donde esta velocidad se descompone (reapecto del sistema de coordenadas tomado), como: 
 $$\overline{v}_0 = (-v_0 \cos{ \frac{\pi}{4} }, -v_0 \sin{ \frac{\pi}{4} }, 0) = -v_0(\frac{\sqrt{2}}{2},\frac{\sqrt{2}}{2}, 0)$$

Entonces, segun las ecuaciónes dinamicas, la cinematica para el proyectil involucra un **M.R.U** en $\hat{x}$ y un **M.R.U.V.** en $\hat{y}$

Cinematica del proyectil $ \left\{ \begin{array}{ccc}
  \hat{x} : & x_p(t) = L  -v_0 \frac{\sqrt{2}}{2} \vartriangle{t} & (5)\\
  \hat{y} : & y_p(t) = v_0 \frac{\sqrt{2}}{2} \vartriangle{t} - \cfrac{\ddot{y_p}}{2} \vartriangle{t}^2 & (6)\\ 
 \end{array}\right.$

(b) Calcule las aceleraciones de ambos cuerpos. Diga para qué valores de $v_0$ el proyectil alcanza a la varilla.
--------------------------------------------------------------------------------
 
Retomemos las ecuaciónes de Newton para la Varilla un momento para ver que nos dicen de la cinematca.
 
De (3) sacamos que:
$$\ddot{y}_v(t) = - g \sin{\alpha} \hspace{2cm} (7)$$
Integramos para obtener la velocidad:
$$
\cfrac{d \dot{y}}{dt} = - g \sin{\alpha} \hspace{0.5cm} \Leftrightarrow \hspace{0.5cm}
\int_{0}^{\dot{y}(t)} d \dot{y} = - g \sin{\alpha} \int_{0}^{t} dt
\hspace{0.5cm} \Leftrightarrow \hspace{0.5cm} \dot{y}(t) = - g \sin{\alpha} t \hspace{2cm} (8)
$$
Integramos otra vez para obtener la ecuación para la posición en y de la varilla:
$$
\cfrac{d y}{dt} = - g \sin{\alpha} t \hspace{0.5cm} \Leftrightarrow \hspace{0.5cm}
\int_{H}^{y(t)} dy = - g \sin{\alpha} \int_{0}^{t} t \cdot dt
\hspace{0.5cm} \Leftrightarrow \hspace{0.5cm} y(t) = y_v(t_0) - g \sin{\alpha} \frac{t^2}{2} \hspace{2cm} (9)
$$
 Luego, la aceleración de el proyectil esta dado por:
$$\ddot{y}_p(t) = - g \sin{\alpha} \hspace{2cm} (10)$$

Con esto tenemos todo para plantear el encuentro entre la varilla y el proyectil. Empecemos pensando en el tiempo de encuentro. El tiempo de encuentro se da cuando la $y_v(t_e) = y_p(t_e)$, además tenemos que $x_p(t_e) = 0$

Planteamos primero esta ultima condición, de la ecuación (5) sacamos que:
$$x_p(t_e) = 0 = L - v_0 \frac{\sqrt2}{2} (t_e - 1) 
\hspace{0.5cm}\Leftrightarrow \hspace{0.5cm}
t_e = \frac{2 L}{v_0 \sqrt{2}} + 1 \hspace{2cm} (11)$$

Tenemos el tiempo de encuentro, ahora como la varilla tiene un cierto largo, la posición de $\hat{y}_p(t_e)$ puede variar entre dos posiciones las cuales son los extremos de la varilla, por lo que tenemos un rango de valores estos va a ser:
$$d - \frac{g \sin{\alpha}t_e}{2}^2 \leq y_p(t_e) \leq H -  \frac{g \sin{\alpha}t_e}{2}^2$$
Evaluando las expreciones en $t_e$ y despejando se obtiene:
$$\cfrac{\sqrt{2}L g \sin{\alpha}}{H-L-\frac{g}{2}\sin{\alpha}} \leq v_0 \leq 
\cfrac{\sqrt{2}L g \sin{\alpha}}{H-L+d - \frac{g}{2}\sin{\alpha}}
\hspace{2cm} (12)$$
