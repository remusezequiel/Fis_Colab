i) Si el plano inclinado esta fijo (es decir, no se mueve) como son las componentes $\hat{x}$ e $\hat{y}$ de la aceleración del bloque fijando el eje $x$ en el piso.

Notemos que, para la masa $m_1$, la normal se descompone como el vector $\overline{N} = (N_x, N_y) = (-N \cos{(\varphi)} , N \sin{(\varphi)})$. 

Por lo que las ecuaciónes de Newton quedan:

$\left\{ \begin{array}{cc} 
      -N \cos{(\varphi)} = m_1 \ddot{x}_1 & (1)\\
      N \sin{(\varphi)} - m_1 g = m_1 \ddot{y}_1 & (2)
\end{array}\right.$

Por otro lado, sobre el plano inclinado actuan $\overline{N_1}$, $\overline{N_2}$ y $\overline{P_2}$. Obteniendo las siguientes ecuaciónes:

$\left\{ \begin{array}{cc} 
      N \cos{(\varphi)} = m_2 \ddot{x}_2 = 0 & (3)\\
     N_2 - N \sin{(\varphi)} - m_2 g = m_2 \ddot{y}_2 = 0 & (4)
\end{array}\right.$

En particular, estas dos ultimas ecuaciónes, en principio, no aportan nada al problema indicado en este inciso. Sin embargo Existe un vinculo entre las aceleraciones de las masas $m_1$ y $m_2$.

Tomemos la posición del bloque $m_1$ en cierto $t \neq 0$ como $(x_1(t), y_1(t))$ y tomemos un punto fijo en el plano $(x_2, y_2)$. Como el punto tomado en el plano es fijo, pues el plano no se mueve. Tenemos que el desplazamiento de la masa respecto al punto fijo esta dado por:

 $\left\{ \begin{array}{cc} 
      D = \cfrac{x_2-x_1(t)}{\cos{(\varphi)}} & (5)\\
      D = \cfrac{y_2-y_1(t)}{\sin{(\varphi)}} & (6)
\end{array}\right.$

Luego, como se debe cumplir $D=D$, llegamos a que $$(y_2 - y_1(t)) = (x2-x_1(t)) \tan{(\varphi)}$$

Ahora, como el punto en el plano es fijo: $\ddot{x}_2 = \ddot{y}_2 = 0$. 
Luego, las aceleraciones de $m_1$ estan relacionadas entre si por la ecuación:

$$\ddot{y}_1) = \ddot{x}_1 \tan{(\varphi)}$$
