Lo que vamos a hacer en esta seccion del COLAB es mostrar mediante una animación las configuraciones del MODO 1 y del MODO 2 de este sistema.
Para eso primero vamos a definir parámetros razonables. 

```
import numpy as np
import matplotlib.pylab as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```
rc('animation', html='html5');
```

```
l0 = 3
fig, ax = plt.subplots();       #en esta celda defino como es el formato de mi figura a la que llamo fig. Siempre que yo grafique alguna curva en un formato fig, me aparecerá como se muestra aca
ax.set_xlabel('$x(t)$');  	    #nombre del eje x
ax.set_xlim((0,9));             #límites del eje x
ax.set_ylim((-2,2));            #límites del eje y
ax.grid(True);                  #True me dice que SI quiero que me muestre la grilla
ax.plot([l0,l0], (-1.5,1.5), '--',color  = 'green');  #esto es para que me grafique las lineas verdes que son los puntos de equilibrio
ax.plot([2*l0,2*l0], (-1.5,1.5), '--',color  = 'green');
ax.text(3.05,1,'$l_0$',color = 'green');              #esto es para que le ponga nombre a las lineas verdes que son los puntos de equilibrio
ax.text(6.05,1,'$2l_0$',color = 'green');
```

```
linea, = ax.plot([],[],'o',ms = 10,color = 'r', label = 'Movimiento particula a',zorder = 3)  #esta es la masa a
lineResa, = ax.plot([],[],ls = '--',color = 'r', lw = 2, zorder = 2)                          #esta es la linea que representa mi primer resorte
lineb, = ax.plot([],[],'o',ms = 10,color = 'b', label = 'Movimiento particula b',zorder = 3)  #esta es la masa b
lineResb, = ax.plot([],[],ls = '--',color = 'b', lw = 2, zorder = 1)                          #esta es la linea que representa mi segundo resorte
```

```
#MODO 1
def MODO1(i):
  A1 = 1 # como estamos solo en un modo elijo una amplitud arbitraria. 
  B1 = 1.5*A1
  phi = 0 # y fase arbitraria
  k = 15 #N/m
  m = 1 #kg
  l0 = 3

  omega1=np.sqrt(k/(2*m))

  t= np.linspace(0,15,100)                  # aca le estamos diciendo que genere equispaciadamente 50 tiempos entre 0 y 5seg
  x_a=A1*np.cos(omega1*t[i] + phi) + l0
  x_b=B1*np.cos(omega1*t[i] + phi) + 2*l0
  linea.set_data(x_a, 0)                    #grafico la posicion x_a 
  lineResa.set_data([0,x_a],[0,0])          #grafico el primer resorte
  lineb.set_data(x_b, 0)                    #grafico la posicion x_b
  lineResb.set_data([0,x_b],[0,0])          #grafico el segundo resorte
  return (linea,)
  
```

```
anim = animation.FuncAnimation(fig, MODO1, frames=100, interval=150)
anim
```

```
#MODO 2
def MODO2(i):
  A2 = 1 # como estamos solo en un modo elijo una amplitud arbitraria. 
  B2 = -A2
  phi2 = 0 # y fase arbitraria
  k = 15 #N/m
  m = 1 #kg
  l0 = 3

  omega2=np.sqrt((3*k)/m)

  t= np.linspace(0,16,100) # aca le estamos diciendo que genere equispaciadamente 50 tiempos entre 0 y 5seg
  x_a=A2*np.cos(omega2*t[i] + phi2) + l0
  x_b=B2*np.cos(omega2*t[i] + phi2) + 2*l0
  linea.set_data(x_a, 0)
  lineResa.set_data([0,x_a],[0,0])
  lineb.set_data(x_b, 0)
  lineResb.set_data([0,x_b],[0,0])
  return (linea,)
```

```
anim = animation.FuncAnimation(fig, MODO2, frames=100, interval=150)
anim
```
